*Exercício 1: Usando o comando print*

Imprima a frase "Olá, mundo!" na tela.

*Exercício 2: Usando f-string para Formatção*

Crie um programa que solicite ao usuário seu nome e, em seguida, imprima uma mensagem de saudação com seu nome usando uma f-string.

*Exercício 3: Usando format() para Formatção*

Crie um programa que solicite ao usuário seu nome e idade. Em seguida, imprima uma mensagem de boas-vindas com esses dados usando a função format()

*Exercício 4: Conversão de Tipos*

Crie um programa que solicite ao usuário um número inteiro e, em seguida, multiplique esse número por 2 e imprima o resultado.

*Exercício 5: Soma de Números*

Crie um programa que solicite ao usuário dois números inteiros, some-os e imprima o resultado.

*Exercício 6: Operadores Aritméticos*

Crie um programa que solicite ao usuário dois números inteiros e, em seguida, realize as seguintes operações:
- Some os números.
- Subtraia o segundo número do primeiro.
- Multiplique os números.


*Exercício 7: Operadores de Comparação Lógica com Número de Referência*

Crie um programa que solicite ao usuário um número inteiro.
Peça ao usuário para inserir um número de referência.
Em seguida, verifique e imprima se o número inserido pelo usuário é:
- Igual ao número de referência.
- Maior que o número de referência.
- Menor que o número de referência.

*Exercício 8: Verificação de Divisibilidade*

Crie um programa que solicita ao usuário um número e, em seguida, imprime mensagens indicando:
- Se o número é divisível por 3.
- Se o número é divisível por 4.
- Se o número é divisível por 3 e por 4.
- Se o número é divisível por 3 ou por 4.

*Exercício 9: Verificação de Paridade*

Crie um programa que solicita ao usuário um número inteiro e, em seguida, imprime se o número é par ou ímpar.

*Exercício 10: Verificação de Maioridade*

Crie um programa que solicita ao usuário sua idade e, em seguida, verifica se a pessoa é maior de idade (idade maior ou igual a 18 anos).

*Exercício 11: Verificação de Número Positivo, Negativo ou Zero*

Crie um programa que solicita ao usuário um número e, em seguida, verifica se o número é positivo, negativo ou igual a zero.

*Exercício 12: Verificação de Número Maior*

Crie um programa que solicita ao usuário dois números inteiros e, em seguida, verifica e imprime qual deles é o maior.

*Exercício 13: Acesso a Elementos de uma Lista*

Crie uma lista de nomes de frutas: "maçã", "banana", "laranja", "uva", "morango".
Use a indexação para acessar e imprimir o terceiro elemento da lista.

*Exercício 14: Modificação de Elementos de uma Lista*

Crie uma lista de números inteiros de 1 a 5.
Use a indexação para modificar o terceiro elemento da lista para 10.
Imprima a lista resultante

*Exercício 15: Extração de Partes de uma Lista*

Crie uma lista de cores: "vermelho", "verde", "azul", "amarelo", "roxo".
Use a indexação para criar uma nova lista que contenha apenas as três primeiras cores.
Imprima a nova lista.

*Exercício 16: Concatenação de Listas*

Crie duas listas: uma com números de 1 a 3 e outra com números de 4 a 6.
Use a indexação para criar uma terceira lista que contenha a concatenação das duas listas.
Imprima a terceira lista.

*Exercício 17: Criação e Acesso a Elementos de uma Tupla*

Crie uma tupla chamada meses com os nomes dos meses do ano.
Acesse e imprima o terceiro mês da tupla.

*Exercício 18: Criação e Acesso a Elementos de uma Tupla*

Crie uma tupla com o nome e a idade de uma pessoa e escreva na tela uma saudação com as informações contidas na tupla

*Exercício 19: Impressão de Números Pares*

*Exercício 20: Soma de Números Ímpares*

*Exercício 22: Contagem de Vogais*

Crie uma string contendo uma frase de sua escolha.

*Exercício 23: Tabuada de Multiplicação*

*Exercício 24: Contagem Crescente*

*Exercício 25: Contagem Regressiva*

*Exercício 27: Jogo de Adivinhação*

*Exercício 28: Contagem de Fatoriais*

*Exercício 29: Vamos criar uma calculadora (simples) com as operações soma, subtração, multiplicação, divisão.*

Dica: crie funções para tarefas específicas como, por exemplo, uma função `soma(a,b)`, uma função `subtracao(a,b)`, ... e depois use estas funções para criar a função `calculadora()`.

Lembre-se de que não se pode dividir por zero.

*Exercício 30: Em um curso da Sirius temos alunos dos mais variados perfis, tendo diferentes faixas etárias, morando em diferentes estados do Brasil e que tiraram diferentes notas. Essas informações estão em diferentes listas, uma somente com os nomes, outras somente com as notas, outra com a idades e outro com os estados de residência.*

Com base nisso crie funções que ajudem a responder as perguntas considerando os dados a seguir.

1) Crie uma função que seleciona o nome de todas as pessoas que residem em um determinado estado. Por exemplo, a função `alunos_por_estado(nomes, estados, 'SP')` deve retornar todos os nomes dos alunos que moram em São Paulo.

2) Crie uma função que calcula quantas pessoas foram aprovadas por estado, considerando média 6. Por exemplo, a função `quant_aprov_estado(estados, notas, 'SP')` deve retornar quantos alunos foram aprovados no estado de São Paulo.

3) Crie uma função que calcula quantos alunos há por estado. Por exemplo, a função `alunos_por_estado(estados)` deve retornar uma mensagem dizendo quantos alunos há em cada estado.

4) Crie uma função que calcula a média das notas dos alunos de um determinado estado. Por exemplo, a função `media_por_estado(notas, estados, 'SP')` deve retornar a média das notas do alunos (com duas casas decimais) do estado de São Paulo.

5) Crie uma função que calcula a média das idades dos alunos aprovados (considerando media 6) de um determinado estado. Por exemplo, a função `media_idade_aprov_por_estado(notas, estados, idades, 'SP')` deve retornar a média das idades do alunos (com duas casas decimais) aprovados que residem no estado de São Paulo.
