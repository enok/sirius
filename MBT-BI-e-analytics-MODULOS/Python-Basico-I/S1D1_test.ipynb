```python
class Curso:
    def __init__(self, nome_curso, instrutores) -> None:
        self.nome_curso = nome_curso
        self.instrutores = instrutores
    
    def imprimir_instrutores(self):
        [print(professor) for professor in self.instrutores]
```

```python
meu_curso = Curso("Python para iniciantes", ["Rodrigo", "Thiago"])
meu_curso.imprimir_instrutores()
```
