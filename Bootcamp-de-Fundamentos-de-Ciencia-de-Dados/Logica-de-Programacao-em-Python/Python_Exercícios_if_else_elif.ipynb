### Exercício 1

Crie um programa que leia a idade do usuário e classifique-o em:
- Criança – 0 a 12 anos
- Adolescente – 13 a 17 anos
- Adulto – acima de 18 anos
 -Se o usuário digitar um número negativo, mostrar a mensagem que a idade é inválida


```python
nome = input('Digite seu nome:   ')
idade = int(input('Digite sua idade:    '))
```

```python
if (idade < 0):
  print(f'Idade inválida')
elif (idade >= 0) and (idade <= 12):
  print(f'{nome} é Criança')
elif (idade >= 13) and (idade <= 17):
  print(f'{nome} é Adolescente')
else:
  print(f'{nome} é Adulto')
```

### Exercício 2

Crie um programa que leia um número inteiro e apresenta as seguintes mensagens:
- “O valor está na faixa permitida”, caso o valor esteja entre 1 e 9
- “O valor está fora da faixa permitida”, caso o valor seja menor que 1 OU maior que 9

```python
numero = int(input('Digite um número:    '))
if (numero >= 1) and (numero <= 9):
  print("O valor está na faixa permitida")
elif (numero < 1) or (numero > 9):
  print("O valor está fora da faixa permitida")
```

### Exercício 3

Crie um programa que calcule a média de um aluno que cursou a disciplina de Estatística, a partir da leitura das notas M1, M2 e M3; passando por um cálculo da média aritmética. Após a média calculada, devemos anunciar se o aluno foi aprovado, reprovado ou pegou exame
- Se a média estiver entre 0.0 e 4.0, o aluno está reprovado
- Se a média estiver entre 4.1 e 6.0, o aluno pegou exame
- Se a média for maior do que 6.0, o aluno está aprovado
- Se o aluno pegou exame, deve ser lida a nota do exame. Se a nota do exame for maior do que 6.0, está aprovado, senão; está reprovado

```python
nota_m1 = float(input('Digite a nota M1:    '))
nota_m2 = float(input('Digite a nota M2:    '))
nota_m3 = float(input('Digite a nota M3:    '))
```

```python
media = (nota_m1 + nota_m2 + nota_m3) / 3
print(f'Média: {media}')
```

```python
if (media >= 0.0) and (media <= 4.0):
  print('O aluno está reprovado')
elif (media >= 4.1) and (media <= 6.0):
  print('O aluno pegou exame')
  nota_exame = float(input('Informe a nota tirada no exame:    '))
  if nota_exame > 6.0:
    print('O aluno está aprovado')
  else:
    print('O aluno está reprovado')
elif (media > 6.0):
    print('O aluno está aprovado')
```

### Exercício 4
Crie um programa que leia dois números reais e uma operação, e com isso imprima o resultado de acordo com a operação escolhida pelo usuário: adição, subtração, multiplicação e divisão.
Caso seja informada operação inválida, mostrar na tela

```python
def soma(a, b):
  print('\nEscolhida a operação de soma')
  return a + b

def subtracao(a, b):
  print('\nEscolhida a operação de subtração')
  return a - b

def multiplicacao(a, b):
  print('\nEscolhida a operação de multiplicação')
  return a * b

def divisao(a, b):
  print('\nEscolhida a operação de divisão')
  if (b == 0):
    return None
  return a / b
```

```python
def calculadora():
  while(True):
    print('\n\nEscolha o número da operação a ser realizada: ')
    print('    1) soma')
    print('    2) subtração')
    print('    3) multiplicação')
    print('    4) divisão')
    print('    5) encerrar calculadora')

    choice = int(input())

    match choice:
      case 1:
        a, b = busca_numeros()
        print(soma(a, b))
      case 2:
        a, b = busca_numeros()
        print(subtracao(a, b))
      case 3:
        a, b = busca_numeros()
        print(multiplicacao(a, b))
      case 4:
        a, b = busca_numeros()
        print(divisao(a, b))
      case 5:
        print('Encerrando calculadora')
        break
      case _:
        print('Opção inválida')

def busca_numeros():
  numero_1 = float(input('Digite o primeiro valor: '))
  numero_2 = float(input('Digite o segundo valor: '))

  return numero_1, numero_2
```
