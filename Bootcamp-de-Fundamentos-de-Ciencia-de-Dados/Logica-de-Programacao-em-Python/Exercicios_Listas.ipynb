##### Exercício 8  -  Programa Python para trocar os primeiros e últimos elementos em uma lista

Dada uma lista, escreva um programa Python para trocar o primeiro e o último elemento da lista.


Exemplos:
```
Input : [12, 35, 9, 56, 24],
Output : [24, 35, 9, 56, 12],

Input : [1, 2, 3],
Output : [3, 2, 1]
```



````python
def swap(list):
    list[0], list[-1] = list[-1], list[0]
````

````python
list = [12, 35, 9, 56, 24]
print(f'lista antes da troca: {list}')
swap(list)
print(f'lista depois da troca: {list}')
````

````python
list = [1, 2, 3]
print(f'lista antes da troca: {list}')
swap(list)
print(f'lista depois da troca: {list}')
````
