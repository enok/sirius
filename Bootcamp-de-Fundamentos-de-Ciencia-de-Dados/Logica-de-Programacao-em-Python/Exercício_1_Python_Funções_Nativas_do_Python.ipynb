# Funções internas

Uma função é uma sequência de comandos que executa alguma tarefa. Veja alguns modelos:

#### Imprime qual o tipo do número 2

```python
print(type(2))
```

#### Imprime qual o valor máximo da string (seguindo a tabelas ASCII), no caso o caractere 'u' (117 em decimal)

```python
print(f'\'{max('Alo mundo')}\'')
```

#### Imprime qual o valor máximo da string (seguindo a tabelas ASCII), no caso o caractere espaço: ' ' (32 em decimal)

```python
print(f'\'{min('Alo mundo')}\'')
```

#### Imprime qual o valor máximo o array, no caso 55

```python
print(max([1,2,4,55]))
```

#### Imprime qual o valor mínimo o array, no caso 1

```python
print(min([1,2,4,55]))
```

#### Imprime qual o tamanho da string, no caso 9

```python
print(len('Alô mundo'))
```

# Funções de conversão de tipo

São funções internas que convertem valores de um tipo para para outro. Veja alguns exemplos (str,int,float):

```python
# converte o numero 32 para inteiro

print(int('32'))
```

```python
# tenta coverter a string 'Hello' para inteiro, o que dá erro por não ser valor numérico
print(int('Hello'))
```

```python
# converte o valor float 5.777 para inteiro, o que na prática remove a parte decimal do número ficando somente 5
print(int(5.777))
```

```python
# tenta coverter a string 'OLA' para inteiro, o que dá erro por não ser valor numérico

print(int('OLA'))
```

```python
# converte o valor float -9.4 para inteiro, o que na prática remove a parte decimal do número ficando somente -9

print(int(-9.4))
```

```python
# converte o número inteiro 32 para float, ficando assim 32.0

print(float(32))
```

```python
# converte o número inteiro 2.589 para float

print(float('2.589'))
```

# Funções matemáticas

Python tem um módulo matemático que oferece a maioria das funções matemáticas comuns. Um módulo é um arquivo contendo definições e comandos em Python para serem usados em outros programas em Python. Há diversos módulos do Python que fazem parte da biblioteca padrão. A ideia é que você pode importar códigos escritos por outras pessoas. Vamos importar um desses módulos.

```python
import math as m
```

```python
# Para acessar uma das funções, é preciso especificar o nome do módulo e o nome da função, separados por um ponto.

# imprime o valor de pi
pi = m.pi
print(pi)
```

```python
# math.comb(n, k) Retorna o número de maneiras de escolher k itens de n itens sem repetição e sem ordem.
# Por exemplo, considere um conjunto com seis elementos que serão tomados dois a dois. Observe que aqui n = 6 e k = 2 logo :

m.comb(6,2)

```

```python
## Quantas combinações na mega-sena---- k=6, n=60
m.comb(60,6)
```

```python
###Composição e chamada de funcoes:

# imprime 8 * raiz quadrada de 16 = 8.0 * 4 = 32.0
x = m.sqrt(16)
y =  8*x
print(y)
```
