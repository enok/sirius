### Exercício 1


Ler 5 notas e informar a média usando loop for


```python
notas = []

for i in range (1, 6):
    notas.append(float(input("Informar a nota: ")))
    print(f'nota: {notas[i-1]}')

print('----------------------------------------')
print(f'média das notas: {sum(notas)  / 5}')
```

### Exercício 2
Ler 5 notas e informar a média usando loop while

```python
notas = []
i = 0
while i < 5:
    notas.append(float(input("Informar a nota: ")))
    print(f'nota: {notas[i]}')
    i += 1

print('----------------------------------------')
print(f'média das notas: {sum(notas)  / 5}')
```

### Exercício 3

Imprimir a tabuada do número 3 (3 x 1 = 1 - 3 x 10 = 30) usando loop for

```python
for i in range (1, 11):
    print(f'3 x {i:2d} = {3*i:2d}')
```
