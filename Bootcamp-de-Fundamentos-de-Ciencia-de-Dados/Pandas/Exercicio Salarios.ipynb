**Qual é a média da  Remuneração base(BasePay) ?**

```python
df['BasePay'] = df['BasePay'].replace('Not Provided', np.nan)
df['BasePay'] = df['BasePay'].astype(float)
print(df['BasePay'].mean())
```

**Qual é a máximo de horas extras(OvertimePay) no conjunto de dados?**

```python
df['OvertimePay'] = df['OvertimePay'].replace('Not Provided', np.nan)
df['OvertimePay'] = df['OvertimePay'].astype(float)
print(df['OvertimePay'].max())
```

**Qual é o cargo de JOANNE HAYES-WHITE'?**

```python
df[df['EmployeeName'] == 'JOANNE HAYES-WHITE']['JobTitle']
```

**Quanto ganha JOANNE HAYES-WHITE (incluindo benefícios)?**

```python
df[df['EmployeeName'] == 'JOANNE HAYES-WHITE']['TotalPayBenefits']
```

**Qual é o nome da pessoa mais bem paga (incluindo benefícios)?**

```python
person = df.sort_values(by=['TotalPayBenefits'], ascending=False).iloc[0]
print(f"TotalPayBenefits: {person['TotalPayBenefits']}")
print(f"EmployeeName: {person['EmployeeName']}")
```

**Qual é o nome da pessoa com o salário mais baixo (incluindo benefícios)? Você percebe algo estranho sobre o quanto ele ou ela é pago?**

```python
person = df.sort_values(by=['TotalPayBenefits']).iloc[0]
print(f"TotalPayBenefits: {person['TotalPayBenefits']}")
print(f"EmployeeName: {person['EmployeeName']}")
```

**Qual foi a média da Remuneração base (BasePay) de todos os funcionários por ano? (2011-2014) ?**

```python
df[['Year', 'BasePay']].groupby(['Year']).mean()
```

**Quais são as 5 profissões mais comuns?**

**Quantos cargos foram representados por apenas uma pessoa em 2013? (ou seja, cargos com apenas uma ocorrência em 2013?)**

```python
df[df['Year'] == 2013]['JobTitle'].value_counts()[1]
```
